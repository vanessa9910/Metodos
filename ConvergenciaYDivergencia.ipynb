```python
import numpy as np
import matplotlib.pyplot as plt

def xnew(x):
    return (2*x**2+3)/5
x0 = 2
x1 = 0
itera = 0
x0Array = np.zeros(100)
x1Array = np.zeros(100)
xexe = np.zeros(100)

for i in range(10):
    x1 = xnew(x0)
    xexe[i] = i
    x0Array[i] = x0
    x1Array[i] = x1
    if abs(x1-x0)< 0.00000001:
        break
    x0 = x1
    itera += 1
    
plt.plot(xexe,x0Array,x1Array)
plt.grid()
```
