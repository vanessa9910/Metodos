```python
#Metodo de Euler
#y' -e^x= 0   y(0)=1   y(5)=?
import math
import numpy as np
from sympy import *
import matplotlib.pyplot as plt


def f(x,y):
    return math.exp(x)


def euler(y0,x,h,f):
    y = []
    y.append(y0)
    for i in range (1,len(x)):
        y.append(y[i-1] + h*f(x[i-1],y[i-1])) 
    return y

n = 20  #puntos
a = 0   #punto inicial
b = 6   #punto final saber el valor
h = abs(a-b)/(n-1)

x = np.linspace(a,b,n)
y = euler(1,x,h,f)

plt.plot(x,y,'g')
plt.plot(x,[math.exp(xi) for xi in x], 'b')
plt.grid()
```
